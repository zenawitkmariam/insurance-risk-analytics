### Check Missing Values


```python
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100
missing_table = pd.concat([missing, missing_percent], axis=1, keys=['Missing Values', '% of Total'])
print(missing_table)
```
